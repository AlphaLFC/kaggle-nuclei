Explore what the data looks like
===

```python
from os.path import join as opj, basename
from glob import glob

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
image_dir = '/home/rlan/datasets/nuclei/stage1/train'
data_loader = (opj(img, 'images', basename(img) + '.png') for img in glob(opj(image_dir, '*')))
```

```python
for img_path in data_loader:
    im = plt.imread(img_path)
    print(img_path)
    print(im.shape)
    plt.figure(figsize=(6, 6))
    plt.imshow(im)
    plt.figure(figsize=(8, 8))
    for c in range(im.shape[2]):
        channel = im[:, :, c]
        print(c, channel.min(), channel.max())
        plt.subplot(2, 2, c + 1)
        plt.imshow(channel, cmap='gray')
    break
```
